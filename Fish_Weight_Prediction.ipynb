```python
Fwp_data = Fwp_data.drop([40],axis=0);
```

```python
Fwp_data.loc[39:41,:]
```

```python
#Encoding Categorical variable
Fwp_duplicate = pd.get_dummies(Fwp_data['Species'],drop_first=True);
```

```python
Fwp_data = Fwp_data.drop('Species',axis=1)
```

```python
Fwpfinals = pd.concat([Fwp_data,Fwp_duplicate],axis=1)
```

```python
Fwpfinals.head()
```

```python
#Train Test Split,Formatting data
x = Fwpfinals.drop('Weight',axis=1)
y = Fwpfinals['Weight'];
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
#build model
from sklearn.lin_mod import LinearRegression, ElasticNet
```

```python
lin_mod = LinearRegression()
```

```python
lin_mod.fit(x_train,y_train)
```

```python
y_pred = lin_mod.predict(x_test)
```

```python
lin_mod.coef_
```

```python
#understad predictions using metrics
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error;
```

```python
Mae = mean_absolute_error(y_test,y_pred)
MSE = mean_squared_error(y_test,y_pred)
RMSE = np.sqrt(MSE)
regressinsqr = r2_score(y_test,y_pred)
```

```python
print(Mae)
print(MSE)
print(RMSE)
print(regressinsqr)
```
